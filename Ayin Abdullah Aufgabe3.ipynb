```python
from PIL import Image
import os
import pickle
#import data
img1 = Image.open("schuetze.jpg").resize((100,100)).convert('L')
img1.putalpha(128)
img2 = Image.open("steinbock.jpg").resize((100,100)).convert('L')
img2.putalpha(128)
img3 = Image.open("wassermann.jpg").resize((100,100)).convert('L')
img3.putalpha(128)
img4 = Image.open("fische.jpg").resize((100,100)).convert('L')
img4.putalpha(128)
img5 = Image.open("widder.jpg").resize((100,100)).convert('L')
img5.putalpha(128)
img6 = Image.open("stier.jpg").resize((100,100)).convert('L')
img6.putalpha(128)
img7 = Image.open("zwilling.jpg").resize((100,100)).convert('L')
img7.putalpha(128)
img8 = Image.open("krebs.jpg").resize((100,100)).convert('L')
img8.putalpha(128)
img9 = Image.open("loewe.jpg").resize((100,100)).convert('L')
img9.putalpha(128)
img10 = Image.open("jungfrau.jpg").resize((100,100)).convert('L')
img10.putalpha(128)
img11 = Image.open("waage.jpg").resize((100,100)).convert('L')
img11.putalpha(128)
img12 = Image.open("skorpion.jpg").resize((100,100)).convert('L')
img12.putalpha(128)

bild=Image.open("meinimag.jpg").resize((800,800))
#img11.show()
```

```python
def Sternzeichen(bild,monat,day):
    if bild.size <((800,800)):
        bild=bild.resize((800,800))
    if monat =='december':
        if day<22:
            bild.paste(img1,(700,0))
        else:
            bild.paste(img2,(700,0))
    if monat =='january':
        if day<20:
            bild.paste(img2,(700,0))
        else:
            bild.paste(img3,(700,0))   
    if monat =='february':
        if day<19:
            bild.paste(img3,(700,0))
        else:
            bild.paste(img4,(700,0))
    if monat =='march':
        if day<21:
            bild.paste(img4,(700,0))
        else:
            bild.paste(img5,(700,0))
    if monat =='april':
        if day<20:
            bild.paste(img5,(700,0))
        else:
            bild.paste(img6,(700,0))   
    if monat =='may':
        if day<21:
            bild.paste(img6,(700,0))       
        else:
            bild.paste(img7,(0,0))   
    if monat =='june':
        if day<21:
            bild.paste(img7,(700,0))    
        else:
            bild.paste(img8,(700,0))   
            
    if monat =='july':
        if day<23:
            bild.paste(img8,(700,0))   
        else:
            bild.paste(img9,(700,0))   
            
    if monat =='august':
        if day<23:
            bild.paste(img9,(700,0))    
        else:
            bild.paste(img10,(700,0))   
    if monat =='september':
        if day<23:
            bild.paste(img10,(700,0))    
        else:
            bild.paste(img11,(700,0))    
    if monat =='october':
        if day<23:
            bild.paste(img11,(700,0))       
        else:
            bild.paste(img12,(700,0))
    if monat =='november':
        if day<22:
            bild.paste(img12,(700,0))
        else:
            bild.paste(img1,(700,0))
    return bild
    


```
